```python
## from data we see that cut and clarity is ordinal feature because it has ranks. for eg. Ideal is highest,poor is lowest.
```

```python
df['cut'].unique()
```

```python
cut_map={
    "Fair":1,"Good":2,"Very Good":3,"Premium":4,"Ideal":5
}
```

```python
df['clarity'].unique()
```

```python
clarity_map = {
    "I1":1,"SI2":2,"SI1":3,"VS2":4,"VS1":5,"VVS2":6,"VVS1":7,"IF":8
}
```

```python
df['color'].unique()
```

```python
color_map= {
    "D":1,"E":2,"F":3,"G":4,"H":5,"I":6,"J":7
}
```

```python
df['cut']=df['cut'].replace(cut_map)
df['clarity']=df['clarity'].map(clarity_map)
df['color']=df['color'].map(color_map)
```
